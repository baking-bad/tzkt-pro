In this chapter we will learn to make basic requests to the TzKT Pro API.

## Initializing client
First thing we should do is creating a client for querying the database.  
TzKT Pro is based on PosgREST project that allows flexible database queries and offer multiple client libraries for various programming languages. The best part is that the query syntax is almost the same across all the implementations which allows you to e.g. use Python tutorials even if you are writing in JS.

```python
from postgrest_py import PostgrestClient
```

```python
client = PostgrestClient("https://pro.tzkt.io")
```

```python
res = await client.from_("Accounts").select("*").limit(5).execute()
```
